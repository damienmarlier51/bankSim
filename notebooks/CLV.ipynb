```python
import matplotlib.pyplot as plt
from pathlib import Path
import hiplot as hip
import seaborn as sns
import pandas as pd
import os
```

```python
notebook_path = os.path.abspath("clv.ipynb")
```

```python
project_dir = Path(notebook_path).resolve().parents[1]
processed_data_dir = "{}/{}".format(project_dir, "data/processed")
```

```python
clv_filepath = "{}/clv.csv".format(processed_data_dir)
```

```python
df = pd.read_csv(clv_filepath)
```

```python
df[["customer", "recency", "T", "frequency", "monetary_value", "pred_revenue"]].sort_values(by="pred_revenue", ascending=False)
```

```python
data = df[["recency", "T", "frequency", "monetary_value", "pred_revenue"]].to_dict(orient="records")
hip.Experiment.from_iterable(data).display()
```

```python
heatmap = plt.scatter(x=df["frequency"].values, y=df["recency"].values, c=df["pred_revenue"], s=3)
plt.xlabel("frequency")
plt.ylabel("recency")
plt.colorbar(heatmap)
```

```python
heatmap = plt.scatter(x=df["monetary_value"].values, y=df["recency"].values, c=df["pred_revenue"], s=3)
plt.xlabel("monetary_value")
plt.ylabel("recency")
plt.colorbar(heatmap)
```

```python
heatmap = plt.scatter(x=df["monetary_value"].values, y=df["frequency"].values, c=df["pred_revenue"], s=3)
plt.xlabel("monetary_value")
plt.ylabel("frequency")
plt.colorbar(heatmap)
```
